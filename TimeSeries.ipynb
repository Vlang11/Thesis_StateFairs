```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime 
import seaborn as sns
from matplotlib.dates import date2num
import matplotlib.dates as mdates
```

```python
#Set up figure with subplots for all years and each individual year 
fig, axes = plt.subplots(4, 2, sharey=True,  sharex=False,figsize=(20,20)) #Set up subplot axes
fig.suptitle('Delaware State Fair Hourly PM2.5 Concentrations and Firework Timing' ,fontsize=20, y= 1.)
fig.tight_layout()

#load in excel file into pandas
xls = pd.ExcelFile(r'C:\Users\vlang\Desktop\Thesis Docs\DelawareStateFairFinal.xlsx')
df1= pd.read_excel(xls, '2019')
df2= pd.read_excel(xls, '2018')
df3= pd.read_excel(xls, '2017')
df4= pd.read_excel(xls, '2016')
df5= pd.read_excel(xls, '2015')
df6= pd.read_excel(xls, '2014')
df7= pd.read_excel(xls, '2013')


def modelrun(df):
    df['date']= pd.to_datetime(df['date'], format='%m/%d/%Y %H:%M')
    df.set_index(df['date'], inplace= True)
    return df

###2019
modelrun(df1)

axes[0,0].set_title("Delaware Hourly Concentrations 2019",y=1.0, pad=-14)
ax1= axes[0,0].plot_date(df1.index, df1['PM2.5'])
axes[0,0].set_ylabel('PM2.5 Concentrations ug/m3')

ax2= axes[0,0].twinx()
ax2.scatter(df1.index, df1['rain'], color='palegreen')

#Control Days
axes[0,0].axvspan(*mdates.datestr2num(['7/13/2019', '7/17/2019']), color='lightsteelblue', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/28/2019 00:00', '8/2/2019']), color='lightsteelblue', alpha=0.5)
#Firework dates
axes[0,0].axvspan(*mdates.datestr2num(['7/18/2019 21:00:00', '7/18/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/19/2019 21:00:00', '7/19/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/20/2019 21:00:00', '7/20/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/22/2019 21:00:00', '7/22/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/23/2019 21:00:00', '7/23/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/24/2019 21:00:00', '7/24/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/26/2019 21:00:00', '7/26/2019 23:00:00']), color='lightcoral', alpha=0.5)
axes[0,0].axvspan(*mdates.datestr2num(['7/27/2019 21:00:00', '7/27/2019 23:00:00']), color='lightcoral', alpha=0.5)


###2018
modelrun(df2)
axes[0,1].set_title("Delaware Hourly Concentrations 2018",y=1.0, pad=-14)
ax1=axes[0,1].plot_date(df2.index, df2['PM2.5'], label='PM2.5 Concentrations')

ax2= axes[0,1].twinx()
ax2.scatter(df2.index, df2['rain'], color='palegreen', label="Precip" )
ax2.set_ylabel('Precip(mm)')

axes[0,1].axvspan(*mdates.datestr2num(['7/14/2018', '7/18/2018 23:00:00']), color='lightsteelblue', alpha=0.5, label='Control Days')
axes[0,1].axvspan(*mdates.datestr2num(['7/29/2018 00:00', '8/3/2018']), color='lightsteelblue', alpha=0.5)

axes[0,1].axvspan(*mdates.datestr2num(['7/19/2018 21:00:00', '7/19/2018 23:00:00']), color='lightcoral', alpha=0.5, label= 'Fireworks')
axes[0,1].axvspan(*mdates.datestr2num(['7/20/2018 21:00:00', '7/20/2018 23:00:00']), color='lightcoral', alpha=0.5)

###2017
modelrun(df3)
axes[1,0].set_title("Delaware Hourly Concentrations 2017",y=1.0, pad=-14)
axes[1,0].plot_date(df3.index, df3['PM2.5'])
axes[1,0].set_ylabel('PM2.5 Concentrations ug/m3')

ax2= axes[1,0].twinx()
ax2.scatter(df3.index, df3['rain'], color='palegreen')

axes[1,0].axvspan(*mdates.datestr2num(['7/15/2017', '7/19/2017 23:00:00']), color='lightsteelblue', alpha=0.5)
axes[1,0].axvspan(*mdates.datestr2num(['7/30/2017 00:00', '8/4/2017']), color='lightsteelblue', alpha=0.5)

axes[1,0].axvspan(*mdates.datestr2num(['7/22/2017 21:00:00', '7/22/2017 23:00:00']), color='lightcoral', alpha=0.5)
axes[1,0].axvspan(*mdates.datestr2num(['7/29/2017 21:00:00', '7/29/2017 23:00:00']), color='lightcoral', alpha=0.5)

###2016
modelrun(df4)
axes[1,1].set_title("Delaware Hourly Concentrations 2016",y=1.0, pad=-14)
ax1=axes[1,1].plot_date(df4.index, df4['PM2.5'])
ax2= axes[1,1].twinx()
ax2.scatter(df4.index, df4['rain'], color='palegreen')
ax2.set_ylabel('Precip(mm)')

axes[1,1].axvspan(*mdates.datestr2num(['7/16/2016', '7/20/2016 23:00:00']), color='lightsteelblue', alpha=0.5)
axes[1,1].axvspan(*mdates.datestr2num(['7/31/2016 00:00', '8/5/2016']), color='lightsteelblue', alpha=0.5)

axes[1,1].axvspan(*mdates.datestr2num(['7/24/2016 21:00:00', '7/24/2016 23:00:00']), color='lightcoral', alpha=0.5)
axes[1,1].axvspan(*mdates.datestr2num(['7/30/2016 21:00:00', '7/30/2016 23:00:00']), color='lightcoral', alpha=0.5)

###2015
modelrun(df5)
axes[2,0].set_title("Delaware Hourly Concentrations 2015",y=1.0, pad=-14)
axes[2,0].plot_date(df5.index, df5['PM2.5'])
axes[2,0].set_ylabel('PM2.5 Concentrations ug/m3')
ax2= axes[2,0].twinx()
ax2.scatter(df5.index, df5['rain'], color='palegreen')


axes[2,0].axvspan(*mdates.datestr2num(['7/17/2015', '7/21/2015 23:00:00']), color='lightsteelblue', alpha=0.5)
axes[2,0].axvspan(*mdates.datestr2num(['8/2/2015 00:00', '8/7/2015']), color='lightsteelblue', alpha=0.5)

axes[2,0].axvspan(*mdates.datestr2num(['7/25/2015 21:00:00', '7/25/2015 23:00:00']), color='lightcoral', alpha=0.5)
axes[2,0].axvspan(*mdates.datestr2num(['7/31/2015 21:00:00', '7/31/2015 23:00:00']), color='lightcoral', alpha=0.5)

###2014
modelrun(df6)
axes[2,1].set_title("Delaware Hourly Concentrations 2014",y=1.0, pad=-14)
axes[2,1].plot_date(df6.index, df6['PM2.5'])

ax2= axes[2,1].twinx()
ax2.scatter(df6.index, df6['rain'], color='palegreen')
ax2.set_ylabel('Precip(mm)')

axes[2,1].axvspan(*mdates.datestr2num(['7/12/2014', '7/16/2014 23:00:00']), color='lightsteelblue', alpha=0.5)
axes[2,1].axvspan(*mdates.datestr2num(['7/27/2014 00:00', '8/1/2014']), color='lightsteelblue', alpha=0.5)

axes[2,1].axvspan(*mdates.datestr2num(['7/22/2014 21:00:00', '7/22/2014 23:00:00']), color='lightcoral', alpha=0.5)
axes[2,1].axvspan(*mdates.datestr2num(['7/26/2014 21:00:00', '7/26/2014 23:00:00']), color='lightcoral', alpha=0.5)

###2013
modelrun(df7)
axes[3,0].set_title("Delaware Hourly Concentrations 2013",y=1.0, pad=-14)
axes[3,0].plot_date(df7.index, df7['PM2.5'])
axes[3,0].set_ylabel('PM2.5 Concentrations ug/m3')
ax2= axes[3,0].twinx()
ax2.scatter(df7.index, df7['rain'], color='palegreen')
ax2.set_ylabel('Precip(mm)')

axes[3,0].axvspan(*mdates.datestr2num(['7/13/2013', '7/17/2013 23:00:00']), color='lightsteelblue', alpha=0.5)
axes[3,0].axvspan(*mdates.datestr2num(['7/28/2013 00:00', '8/2/2013']), color='lightsteelblue', alpha=0.5)

fig.delaxes(axes[3,1])
#figure.tight_layout(pad=3.0)

handles,labels = [],[]
for ax in fig.axes:
    for h,l in zip(*ax.get_legend_handles_labels()):
        handles.append(h)
        labels.append(l)

fig.legend(handles,labels, loc= 'upper right')

plt.show()
```
