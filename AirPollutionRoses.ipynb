```python
from windrose import WindroseAxes
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import numpy as np
import pandas as pd
import matplotlib.ticker as mtick
```

```python
#file = 'C:\Users\vlang\Desktop\Papers\2022_MastersThesis\Files\IowaStateFairTest.xlsx'
wd_data = pd.read_excel(r'C:\Users\vlang\Desktop\Papers\2022_MastersThesis\Files\IowaStateFairTest.xlsx', sheet_name='AllFest')
```

```python
wd_data= wd_data.loc[(wd_data['ws']>=0) & (wd_data['wd'] >=0) & (wd_data['pm25'] >=0)]
df= wd_data.dropna(how='all')
ws= wd_data['ws']
wd = wd_data['wd']
pm25= wd_data['pm25']
mask = ~np.isnan(ws) & ~np.isnan(wd) & ~np.isnan(pm25)
```

```python
ax= WindroseAxes.from_ax()
ax.bar(wd,pm25, normed=True, bins=np.arange(0,36,5),nsector=36, opening=.9, cmap=plt.get_cmap('Reds'))   
ax.set_xticklabels(['E', 'NE', 'N', 'NW',  'W', 'SW', 'S', 'SE'])
ax.quiver(0,0,-.104528,.9945, color='black', angles="xy", scale_units='xy', scale=.15)
fmt = '%.0f%%' 
yticks = mtick.FormatStrFormatter(fmt)
ax.set_yticks(np.arange(0, 8, step=1))
ax.set_yticklabels(np.arange(0, 8, step=1))
ax.yaxis.set_major_formatter(yticks)

plt.title('Iowa State Fair Festival Days 2017-2019', fontsize=18)
ax.set_legend().set_title('Hourly PM2.5 \nConcentrations')
```
